```python
df.drop(columns=['Booking_ID'], inplace=True)
```

```python
input_df=df.drop('booking_status',axis=1)
output_df=df['booking_status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size = 0.2, random_state = 42)
```

```python
x_train['type_of_meal_plan'].mode()[0]

x_train['type_of_meal_plan'].fillna('Meal Plan 1', inplace=True)
x_test['type_of_meal_plan'].fillna('Meal Plan 1', inplace=True)

```

```python
x_train['required_car_parking_space'].mode()[0]

x_train['required_car_parking_space'].fillna('0', inplace=True)
x_test['required_car_parking_space'].fillna('0', inplace=True)
```

```python
boxplot = x_train.boxplot(column=['avg_price_per_room']) 
plt.show()
```

```python
x_train['avg_price_per_room'].median()
```

```python
x_train['avg_price_per_room'].fillna(x_train['avg_price_per_room'].median(), inplace= True)
x_test['avg_price_per_room'].fillna(x_train['avg_price_per_room'].median(), inplace= True)
```

```python
x_train.isna().sum()
```

```python
x_test.isna().sum()
```

```python
x_train.head(20)
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['type_of_meal_plan','room_type_reserved', 'market_segment_type']

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoder.fit(x_train[categorical_cols])

x_train_encoded = encoder.transform(x_train[categorical_cols])
x_test_encoded = encoder.transform(x_test[categorical_cols])

encoded_cols = encoder.get_feature_names_out(categorical_cols)

x_train_encoded_df = pd.DataFrame(x_train_encoded, columns=encoded_cols, index=x_train.index)
x_test_encoded_df = pd.DataFrame(x_test_encoded, columns=encoded_cols, index=x_test.index)

x_train = pd.concat([x_train.drop(columns=categorical_cols), x_train_encoded_df], axis=1)
x_test = pd.concat([x_test.drop(columns=categorical_cols), x_test_encoded_df], axis=1)

```

```python
x_train.info()
```

```python
x_train.head()
```

```python
x_train['required_car_parking_space'] = pd.to_numeric(x_train['required_car_parking_space'], errors='coerce')
x_test['required_car_parking_space'] = pd.to_numeric(x_train['required_car_parking_space'], errors='coerce')
```

```python
x_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y_train =encoder.fit_transform(y_train)
y_test =encoder.fit_transform(y_test)

y_train = pd.Series(y_train)
y_test = pd.Series(y_test)
```

```python
y_train.head(10)
```

# Modelling 

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

RF_class = RandomForestClassifier(criterion= 'gini',max_depth=4)
RF_class.fit(x_train, y_train)
```

```python
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

XGBoost_class = XGBClassifier()
XGBoost_class.fit(x_train, y_train)
```

```python
y_predict_RF=RF_class.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print('\nClassification Report RF\n')
print(classification_report(y_test, y_predict_RF, target_names=['Cancelled','Not_Cancelled']))
```

```python
y_predict_XG=XGBoost_class.predict(x_test)
```

```python
print('\nClassification Report XGBoost\n')
print(classification_report(y_test, y_predict_XG, target_names=['Cancelled','Not_Cancelled']))
```
